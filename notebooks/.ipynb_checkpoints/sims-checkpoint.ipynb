```julia
using DrWatson
@quickactivate projectdir()
```

```julia
include(srcdir("rdpg.jl"))
import Main.rdpg
using PersistenceDiagrams, Pipe, Plots, ProgressMeter, Random, Ripserer, Statistics, StatsBase
```

```julia
function generate_sbm(n, k, p, r)
    f = (x, y) -> r + p * (x == y)
    Z = rand(1:k, n)
    return rdpg.Adjacency(f, Z)
end

function scale(X)
    return StatsBase.standardize(ZScoreTransform, X, dims=1)
end

function subsample(X, a=1.0)
    sample(X |> rdpg.m2t, round(Int, size(X, 1)^a ), replace=false )
end

function diagram(Xn; dim_max=0)
    ripserer(Xn |> Alpha, dim_max=dim_max)
end
```

```julia
p, r = 0.5, 0.05
clust = 3
repeats = 10
order = 0
d=2

N = [200, 500, 1000, 2000, 5000, 10000]
ϵ = [0.5, 2, 5]
n = length(N);
```

```julia
with_eps    = [zeros(repeats, n) for _ in 1:length(ϵ)];
without_eps = [zeros(repeats, n) for _ in 1:length(ϵ)];

prog = Progress(convert(Int, n * repeats * length(ϵ)))

for i in 1:n
    
    A = generate_sbm(N[i], clust, p, r)
    X, _, _ = rdpg.spectralEmbed(A, d=d)
    X_norm = scale(X)
    DX = diagram(X |> subsample, dim_max=order)
    DX_norm = diagram(X_norm |> subsample, dim_max=order)

    for j in eachindex(ϵ), k in 1:repeats
            
        A1 = rdpg.edgeFlip(A, ϵ=ϵ[j])            
        X1, _ = rdpg.spectralEmbed(A1, d=d)
        X1_norm = scale(X1)
        D1_norm = diagram(X1_norm |> subsample, dim_max=order)
        without_eps[j][k, i] = Bottleneck()(DX_norm, D1_norm)

        A2 = (A1 .- rdpg.τ(ϵ[j])) ./ rdpg.σ(ϵ[j])^2
        X2, _ = rdpg.spectralEmbed(A2, d=d)
        D2 = diagram(X2 |> subsample, dim_max=order)
        with_eps[j][k, i] = Bottleneck()(DX, D2)
            
        next!(prog)
    end
end
```

```julia
plt1 = plot(title="ϵ publicly available")
for k in 1:length(ϵ)
    plot!(plt1, N,
        map(x -> x == Inf ? missing : x, with_eps[k]) |> eachcol .|> skipmissing .|> mean,
        marker=:o, ylim=(0, 1),
        label="k=$(ϵ[k])",
        lw=3, fillapha=0.01,
    )
end
plt1
```

```julia
plt2 = plot(title="ϵ not publicly available")
for k in 1:length(ϵ)
    plot!(plt2, N,
        map(x -> x == Inf ? missing : x, without_eps[k]) |> eachcol .|> skipmissing .|> mean,
        marker=:o, ylim=(0, 1),
        label="ϵ=$(ϵ[k])",
        lw=3, fillapha=0.01,
    )
end
plt2
```

```julia
plot(plt1, plt2, size=(800, 400))
```
