```julia
using DrWatson
```

```julia
include(srcdir("rdpg.jl"))
using Main.rdpg
using StatsBase, Pipe, Graphs, GraphIO, LightGraphs, DelimitedFiles, Random
using ProgressMeter, DataFrames
using Plots, Ripserer, PersistenceDiagrams, PersistenceDiagramsBase
using Distances, LinearAlgebra, UMAP, TSne
using SparseArrays
```

```julia
function read_graph(; path, delim='\t', labels=nothing)
    data = Int.(readdlm(path, delim))
    if labels !== nothing
        rownames = labels[:, 1]
        n = length(unique(data))
        indx = map(i -> findall(j -> j == i, rownames), data)
        A = sparse(indx[:, 1], indx[:, 2], Int(1), n, n)
    else
        tmp = unique(data)
        n = length(tmp)
        data .= minimum(tmp) != 1 ? data .+ 1 : data
        A = sparse(data[:, 1], data[:, 2], Int(1), n, n)
    end
    return A |> LightGraphs.LinAlg.symmetrize
end
```

```julia
dim = 100
n = 20000
subsample = true
path_to_graph = datadir("email-Eu-core.txt")
path_to_labels = datadir("email-Eu-core-department-labels.txt")
```

```julia
labels = Int.(readdlm(path_to_labels))
deps = labels[:, 2]
Adjacency = read_graph(path=path_to_graph, delim=' ', labels=nothing)
N = size(Adjacency, 1)
rho = sum(Adjacency) / (N * (N-1))
umap_params = (; n_neighbors=25, metric=Euclidean(), n_epochs=1000, neg_sample_rate=50, repulsion_strength=5)
```

----

```julia
A = copy(Adjacency)
Xhat, _ = rdpg.spectralEmbed(A, d=dim, scale=false);
```

```julia
Xnh = Xhat
embedding_umap_x = umap(Xnh', 2; umap_params...)';
# embedding_umap_x = tsne(Xnh, 2, 0, 2000, 5.0);
```

```julia
plt_umap_x = scatter(
    embedding_umap_x |> rdpg._Matrix_to_ArrayOfTuples,
    ms=1.5, msw=0, legend=false, ratio=1, size=(350,300),
    group=map(x -> x % 5, labels[:, 2]),
    # group=labels[:, 2],
    title="ϵ = ∞"
)
```

----

```julia
Random.seed!(2022)
ϵ = log(N) / (sqrt(N) * rho)
B = (rdpg.edgeFlip(A, ϵ=ϵ) .- rdpg.τ(ϵ)^2) ./ rdpg.σ(ϵ)^2
Yhat, _ = rdpg.spectralEmbed(B, d=dim, scale=false)
Ynh1 = Yhat;
```

```julia
embedding_umap_y1 = umap(Ynh1', 2; umap_params...)';
```

```julia
plt_umap_y1 = scatter(
    embedding_umap_y1 |> rdpg._Matrix_to_ArrayOfTuples,
    ms=1.5, msw=0, legend=false, ratio=1, size=(350,300),
    group=map(x -> x % 5, labels[:, 2]),
    # group=labels[:, 2],
    title="ϵ ≍ polylog(n) ÷ √nρₙ"
)
```

----

```julia
Random.seed!(2022)
ϵ = log(N) / (N * rho)
B = (rdpg.edgeFlip(A, ϵ=ϵ) .- rdpg.τ(ϵ)^2) ./ rdpg.σ(ϵ)^2
Yhat, _ = rdpg.spectralEmbed(B, d=dim, scale=false)
Ynh2 = Yhat;
```

```julia
embedding_umap_y2 = umap(Ynh2', 2; umap_params...)';
```

```julia
plt_umap_y2 = scatter(
    embedding_umap_y2 |> rdpg._Matrix_to_ArrayOfTuples,
    ms=1.5, msw=0, legend=:bottomright, ratio=1, size=(350,300),
    group=map(x -> x % 5, labels[:, 2]),
    # group=labels[:, 2],
    title="ϵ ≍ polylog(n) ÷ nρₙ"
)
```
